```python
import pandas as pd
import numpy as np
import time
import os
import io
from PIL import Image
import matplotlib.pyplot as plt
import pickle
from datetime import timedelta
```

```python
file_dir = "../Processed_Gases_Official_Data/"
Delhi_Stations_name = []
for x in os.listdir(file_dir):
    if "Delhi_" in x:
        Delhi_Stations_name.append(x)
```

```python
coods = pd.read_csv("../station_coordinates.csv", index_col = 0)
coods
```

```python
Delhi_Stations_Coods = coods[coods["Station"].str.contains(", Delhi")]
Delhi_Stations_Coods
```

```python
Station_Values = {}
for name in Delhi_Stations_Coods["Station"]:
    temp = pd.read_csv(file_dir+name+".csv", index_col=0).iloc[:-1]
    temp.set_index('To_Date', inplace=True)
    temp.index = pd.to_datetime(temp.index, format="%d-%m-%Y %H:%S")
    temp = temp.resample('14D').apply(np.mean)
    Station_Values[name] = temp.copy()
```

```python
import folium
def draw_Circle(loc, radius, map, color):
    folium.Circle(
        location=loc,
        radius=int(radius),
        color=color,
        stroke=False,
        fill=True,
        fill_opacity=0.6,
        opacity=1,
    ).add_to(map)
def draw_cross(loc, map):
    mak = folium.map.Marker(loc, icon=folium.features.CustomIcon(
                                            "https://uxwing.com/wp-content/themes/uxwing/download/checkmark-cross/cross-icon.png", 
                                            icon_size=(10, 10)))
    map.add_child(mak)
def get_concat_h(im1, im2):
    dst = Image.new('RGB', (im1.width + im2.width, im1.height))
    dst.paste(im1, (0, 0))
    dst.paste(im2, (im1.width, 0))
    return dst
def draw_bar(x, max_x, date, type_polution):
    px = 1/plt.rcParams['figure.dpi']  # pixel in inches
    plt.subplots(figsize=(100*px, 600*px))
    plt.title(f"{type_polution}\n{date.date()}\n\n")
    plt.bar(0, x)
    plt.ylim(0, max_x*1.05)
    ax = plt.gca()
    ax.xaxis.set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight')
    im = Image.open(buf)
    return im
def draw_star(loc, map):
    mak = folium.map.Marker(loc, icon=folium.features.CustomIcon(
                                            "https://cdn-icons-png.flaticon.com/512/1828/1828884.png", 
                                            icon_size=(30, 30)))
    map.add_child(mak)
```

```python
API_values = pd.read_csv("../Data_Openweathermap/raw_data/Delhi_raw.csv", index_col = 0)
```

```python
API_values.set_index(pd.to_datetime(API_values["Date"]), inplace=True)
API_values.rename(columns={"PM2_5": "PM2.5"}, inplace=True)
API_values = API_values.rolling(14).mean()
```

```python
API_values
```

```python
scale = {}
for type_polution in ["CO", "NO2", "SO2", "PM2.5"]:
    scale[type_polution] = 2000/np.nanmean([x[type_polution].mean() for x in Station_Values.values()])
```

```python
def draw_gif(type_polution, color, prefix=""):
    m = folium.Map(location=[28.65, 77.15])
    list_img = []
    max_polution = API_values[type_polution].max()
    indexes = list(Station_Values.values())[0].index
    for date in indexes:
        print(type_polution, date.date(), "{}/{}".format((len(list_img)+1), len(indexes)))
        m = folium.Map(location=[28.65, 77.15], zoom_start=10)
        draw_star([28.564, 77.145], m)
        for _, row in Delhi_Stations_Coods.iterrows():
            try:
                value = Station_Values[row["Station"]].loc[date][type_polution]
            except:
                value = np.nan

            cood = row[["Latitude", "Longitude"]].to_list()
            if np.isnan(value):
                draw_cross(cood, m)
            else:
                draw_Circle(cood, value*scale[type_polution], m, color)
        if date in API_values.index:
            api_value = API_values[type_polution].loc[date]
        else:
            api_value = 0
        api_bar = draw_bar(api_value, max_polution, date, type_polution)
        img_data = m._to_png()
        list_img.append(get_concat_h(api_bar, Image.open(io.BytesIO(img_data)).crop((400, 100, 1000, 700))))
        plt.close()
    if prefix:
        prefix += "_"
    list_img[0].save(f'{prefix}{type_polution}.gif', save_all=True, append_images=list_img[1:], duration=100)
    with open(f"{prefix}{type_polution}_gif_data.pkl", "wb") as f:
        pickle.dump(list_img, f)
```

```python
colors = {}
colors["SO2"] = "#5e312d"
colors["CO"] = "#807c1f"
colors["PM2.5"] = "#3d3d35"
colors["NO2"] = "#415073"
```

```python
%matplotlib agg

for type_polution in ["CO", "NO2", "SO2", "PM2.5"]:
    start_time = time.time()
    print(type_polution, "started")
    
    draw_gif(type_polution, colors[type_polution])
    
    end_time = time.time()
    print(type_polution, "finished, time used:", str(timedelta(seconds=int(end_time-start_time))))
    print("#"*30)
```
