```python
from os import listdir
from os.path import isfile, join
mypath = '/Users/code/Documents/GitHub/Capstone_MacroX/Processed_Gases_Official_Data/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
settings = EfficientFCParameters()
```

```python
def features_extraction(col_name):
    start = 0
    print(col_name)
    for files in onlyfiles:
        cur_df = pd.read_csv(mypath+files)
        #Define a subset of the dataset
        df_PM  = cur_df.filter([col_name], axis=1).copy()
        # Define scaler to set values between 0 and 1
        df_knn = pd.DataFrame(df_PM, columns = df_PM.columns)
        # Define KNN imputer and fill missing values
        df_knn['id']= files
        cur_mean = df_knn[col_name].mean()
        df_knn = df_knn.fillna(cur_mean)
        if start == 0:
            try:
                features = extract_features(df_knn, column_id='id',default_fc_parameters=settings)
                start += 1
            except:
                continue
        try:
            cur_features = extract_features(df_knn, column_id='id',default_fc_parameters=settings)
            features = pd.concat([features, cur_features], axis=0)
        except:
            continue
    return features
    
```

```python
gases = ['PM2.5','SO2','NO2','CO']
stations = pd.DataFrame(onlyfiles) 
stations.columns = ['file_name']
```

```python

for gas in gases:
    cur_features = features_extraction(gas)
    stations = stations.merge(cur_features,how = 'left',left_on='file_name',right_index=True)
        
```
