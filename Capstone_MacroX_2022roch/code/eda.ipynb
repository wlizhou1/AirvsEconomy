```python
import pandas as pd
import os
```

```python
file_dir = r"E:\UR_Files\UR_Cap_Stone\github\Capstone_MacroX\Processed_Gases_Official_Data"
i = 0
for filename in os.listdir(file_dir):
    if i == 0:
        df = pd.read_csv(file_dir + "\\" + filename)
        i = 1
    else:
        df = pd.concat([df,pd.read_csv(file_dir + "\\" + filename)],ignore_index=True)

df.head()
df.to_csv(r'E:\UR_Files\UR_Cap_Stone\github\Capstone_MacroX\Processed_Gases_Official_Data\full_station.csv',index=False)
```

```python
na_drop_df = df.dropna()
average = na_drop_df.groupby('station').mean()
average
```

```python
from tkinter import font
from matplotlib import pyplot as plt
fig = plt.figure(figsize = (20, 50))
plt.barh(average.index,average['PM2.5'])
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.title('Average PM2.5 Emission',fontsize= 30)
plt.savefig('average_PM25.jpg')
```

```python
fig = plt.figure(figsize = (20, 50))
plt.barh(average.index,average['SO2']) 	#NO2	CO
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.title('Average SO2 Emission',fontsize= 30)
plt.savefig('average_SO2.jpg')
```

```python
fig = plt.figure(figsize = (20, 50))
plt.barh(average.index,average['NO2']) 	#NO2	CO
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.title('Average NO2 Emission',fontsize= 30)
plt.savefig('average_NO2.jpg')
```

```python
fig = plt.figure(figsize = (20, 50))
plt.barh(average.index,average['CO']) 	#NO2	CO
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.title('Average CO Emission',fontsize= 30)
plt.savefig('average_CO.jpg')
```
