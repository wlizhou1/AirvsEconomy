```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Openweathermap/raw_data/Mumbai_raw.csv", index_col = 0)
```

```python
df
```

```python
plt.figure(figsize=(16,10))
df[["Date","CO"]].plot(x="Date", ax=plt.gca())
```

```python
temp = df[["Date","CO"]].copy()
```

```python
temp["CO_ma_7"] = temp["CO"].rolling(7).mean()
temp["CO_ma_30"] = temp["CO"].rolling(30).mean()
```

```python
plt.figure(figsize=(10, 6))
temp.plot(x="Date", ax=plt.gca(), color = ["lightgray", "orange", "red"])
```
