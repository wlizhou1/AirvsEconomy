```python
import pandas as pd
import numpy as np

satellite_NO2 = pd.read_csv(r'E:\UR_Files\UR_Cap_Stone\github\Capstone_MacroX\satellite_data\satellite_data_NO2_India.csv')
```

```python
satellite_NO2['date']=pd.to_datetime(satellite_NO2['date'],errors='coerce')
satellite_NO2=satellite_NO2.sort_values(by='date')
```

```python
NO2_18_21 = satellite_NO2[satellite_NO2.date.between('2018-01-01','2021-12-31')].replace(-1.27E+30,np.NaN)
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler

#Define a subset of the dataset
df_knn = NO2_18_21.filter(NO2_18_21.columns[2:].tolist(), axis=1).copy()

# Define scaler to set values between 0 and 1
df_knn = pd.DataFrame(df_knn, columns = df_knn.columns)

# Define KNN imputer and fill missing values
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
df_knn_imputed_NO2 = pd.DataFrame(knn_imputer.fit_transform(df_knn), columns=df_knn.columns)
df_knn_imputed_NO2.insert(loc=0, column='date', value=NO2_18_21.date.tolist())

df_knn_imputed_NO2

df_knn_imputed_NO2.to_csv(r'E:\UR_Files\UR_Cap_Stone\github\Capstone_MacroX\processed_satellite_data\processed_satellite_NO2.csv', index=False)
```
