```python
import os
```

```python
stations = []
```

```python
stations
```

```python
for station in [x.split("_")[-1] for x in os.listdir("Processed_Gases_Official_Data")]:
    if "-" in station:
        stations.append("-".join(station.split("-")[:-1]))
    else:
        stations.append(".".join(station.split(".")[:-1]))
```

```python
stations
```

```python
locs = [geolocator.geocode(x) for x in stations]
```

```python
locs
```

```python
import pandas as pd
df = pd.DataFrame(columns = ["Station", "Longitude", "Latitude"])
```

```python
df["Station"] = [x.split(".csv")[0] for x in os.listdir("Processed_Gases_Official_Data")]
```

```python
df["Longitude"] = [x.longitude if x!=None else None for x in locs]
df["Latitude"] = [x.latitude if x!=None else None for x in locs]
```

```python
df
```

```python
df.to_csv("station_coordinates.csv")
```
