```python
import pandas as pd
import os
```

```python
open_weather_dir = 'E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/Data_Openweathermap/raw_data/Delhi_raw.csv'
open_weather_Delhi = pd.read_csv(open_weather_dir) 
Delhi_20_21 = open_weather_Delhi[open_weather_Delhi.Date.between('2020-11-24','2021-12-31')]
```

```python
sate_NO2_dir = 'E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/processed_satellite_data/processed_satellite_NO2.csv'
sate_NO2 = pd.read_csv(sate_NO2_dir)
sate_NO2_20_21 = sate_NO2[sate_NO2.date.between('2020-11-24','2021-12-31')]
sate_NO2_20_21['NewDelhi']
# sate_NO2[sate_NO2['['NewDelhi']
```

```python
sate_SO2_dir = 'E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/processed_satellite_data/processed_satellite_SO2.csv'
sate_SO2 = pd.read_csv(sate_SO2_dir)
sate_SO2_20_21 = sate_SO2[sate_SO2.date.between('2020-11-24','2021-12-31')]
missing_sate_SO2_date = list(set(Delhi_20_21.Date)-set(sate_SO2_20_21.date))
missing_sate_SO2_date
```

```python
missing_sate_SO2_date_df =  pd.DataFrame(columns = sate_SO2.columns)
missing_sate_SO2_date_df['date'] = missing_sate_SO2_date
sate_SO2_20_21 = sate_SO2_20_21.append(missing_sate_SO2_date_df)
sate_SO2_20_21=sate_SO2_20_21.sort_values(by='date')
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler

#Define a subset of the dataset
df_knn = sate_SO2_20_21.filter(sate_SO2_20_21.columns[1:].tolist(), axis=1).copy()

# Define scaler to set values between 0 and 1
df_knn = pd.DataFrame(df_knn, columns = df_knn.columns)

# Define KNN imputer and fill missing values
knn_imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
df_knn_imputed_SO2 = pd.DataFrame(knn_imputer.fit_transform(df_knn), columns=df_knn.columns)
df_knn_imputed_SO2.insert(loc=0, column='date', value=sate_SO2_20_21.date.tolist())
sate_SO2_20_21 = df_knn_imputed_SO2.copy()
```

```python
sate_SO2_20_21
```

```python
Delhi_NO2 = pd.DataFrame(columns = ['date','open_NO2','sate_NO2'])
Delhi_NO2['open_NO2'] = list(Delhi_20_21['NO2'])
Delhi_NO2['date'] = list(Delhi_20_21['Date'])
Delhi_NO2['sate_NO2'] = list(sate_NO2_20_21['NewDelhi'])

Delhi_SO2 = pd.DataFrame(columns = ['date','open_SO2','sate_SO2'])
Delhi_SO2['open_SO2'] = list(Delhi_20_21['SO2'])
Delhi_SO2['date'] = list(Delhi_20_21['Date'])
Delhi_SO2['sate_SO2'] = list(sate_SO2_20_21['NewDelhi'])
```

```python
Delhi_NO2
```

```python
file_dir = "E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/Missing Value/Cluster Fill/" 
DelhiList = []
for filename in os.listdir(file_dir):
    if filename.split('_')[1] == 'Delhi':
        DelhiList.append(filename)
for filename in DelhiList:
    cur_df = pd.read_csv(file_dir+filename)
    cur_df['date'] = pd.to_datetime(cur_df['From_Date'])
    cur_df_20_21 = cur_df[cur_df.date.between('2020-11-24','2021-12-31')]
    missing_date = []
    if len(cur_df_20_21) == len(Delhi_20_21):
        station = filename.split('_')[2].split('.')[0]
        Delhi_NO2[station] = list(cur_df_20_21['NO2'])
        Delhi_SO2[station] = list(cur_df_20_21['SO2'])
    else:
        missing_date.append(filename)
```

```python
save_dir = 'E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/combine_data_cluster/'
Delhi_SO2.to_csv(save_dir+'Delhi_SO2.csv',index=False)
Delhi_NO2.to_csv(save_dir+'Delhi_NO2.csv',index=False)
```
