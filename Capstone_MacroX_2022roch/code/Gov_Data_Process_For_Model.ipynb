```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import MinMaxScaler
def process_for_model(filename,tmp_save_dir):
    df=pd.read_csv(filename)
   
    df = df.dropna(axis=1, how='all', inplace=False)
    df.insert(loc=0, column='date', value=pd.to_datetime(df['From_Date'],errors='coerce'))
    df = df.drop(columns=['Unnamed: 0'])
    
    df_18_21 = df[df.date.between('2018-01-01','2022-12-31')]

    df_knn = df_18_21.filter(df_18_21.columns[3:].tolist(), axis=1).copy()
    df_knn = pd.DataFrame(df_knn, columns = df_knn.columns)
    knn_imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
    df_knn = pd.DataFrame(knn_imputer.fit_transform(df_knn), columns=df_knn.columns)
    df_knn.insert(loc=0, column='date', value=df_18_21.date.tolist())
    if not os.path.exists(tmp_save_dir):
        print("temp_save_dir not found, new folder creating")
        os.mkdir(tmp_save_dir)
        print("temp_save_dir created")

    # save csv file
    
    df_knn.to_csv(f"{tmp_save_dir}{filename.replace(file_dir,'')}")
    print("CSV file saved\n"+"="*40)

```

```python
temp_dir = "E:/UR_Files/UR_Cap_Stone/github/Capstone_MacroX/Processed_Gases_Official_Data_For_Model/"
for filename in os.listdir(file_dir):  
    print(filename)
    process_for_model(file_dir+filename, temp_dir)
```
